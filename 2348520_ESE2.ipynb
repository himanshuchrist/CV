## Question 1- Median and Minimum filter

### Loading the image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('/Users/himanshugulechha/Desktop/MSc AI&ML/Sem3/CV/set-1-q1.tif',0)
```

### Using user defined function

```python
m,n=img.shape
mid=np.zeros([m,n])
min=np.zeros([m,n])
for i in range(1,m-1):
    for j in range(1,n-1):
        temp=[img[i-1,j-1],
              img[i-1,j],
              img[i-1,j+1],
              img[i,j-1],
              img[i,j],
              img[i,j+1],
              img[i+1,j-1],
              img[i+1,j],
              img[i+1,j+1]]
        temp=sorted(temp)
        mid[i,j]=temp[4]
        min[i,j]=temp[0]
plt.figure(figsize=(10, 5))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(mid, cmap='gray')
plt.title('Median filter')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(min, cmap='gray')
plt.title('Minimum Filter')
plt.axis('off')
```

### Using built-in function

```python
from scipy.ndimage import median_filter, minimum_filter
mid1=median_filter(img,3)
min1=minimum_filter(img,3)
plt.figure(figsize=(10, 5))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(mid1, cmap='gray')
plt.title('Median filter')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(min1, cmap='gray')
plt.title('Minimum Filter')
plt.axis('off')
```

The adaptive median filter is an extension of the median filter that adjusts its window size dynamically based on the local characteristics of the image. It is particularly useful for images with varying levels of noise, where a fixed window size may not be optimal for noise reduction. The adaptive median filter typically operates in multiple stages, gradually increasing the window size until a suitable level of noise reduction is achieved.

Min filter is useful for finding the darkest points in an image as it replaces each pixel's value with the minimum value within its neighborhood, the filter effectively highlights the darkest points in the image. For eroding light regions adjacent to dark areas the intensity of pixels in light regions adjacent to dark areas is decreased. This is because the minimum value within the neighborhood is assigned to each pixel, leading to a gradual erosion of light regions.

### LoG- Laplacian of Gaussian

```python
img1=cv2.imread('/Users/himanshugulechha/Desktop/MSc AI&ML/Sem3/CV/set-1-q2.tif',0)
image=cv2.GaussianBlur(img1,(3, 3),0)
filtered_image = cv2.Laplacian(image,cv2.CV_8U,ksize=3)
plt.figure(figsize=(10, 5))
plt.subplot(1,3,1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(image, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(filtered_image, cmap='gray')
plt.title('Laplacian of Gaussian')
plt.axis('off')
```

### Canny

```python
canny=cv2.Canny(img1,100,100)
canny1=cv2.Canny(filtered_image,100,100)
plt.figure(figsize=(10, 5))
plt.subplot(1,3,1)
plt.imshow(img1, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(canny, cmap='gray')
plt.title('Canny of Image')
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(canny1, cmap='gray')
plt.title('Canny of Filtered Image')
plt.axis('off')
```
