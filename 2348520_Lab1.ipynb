```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
image = cv2.imread('/Users/himanshugulechha/Desktop/MSc AI&ML/Sem3/CV/Dubai_Marina_Skyline.jpg')
plt.imshow(image)
print(image.shape)
```

```python
kernel1 = np.array([[0, 0, 0], 
                    [0, 1, 0], 
                    [0, 0, 0]]) 
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1) 
plt.imshow(img)
print(img.shape)
```

```python
kernel2 = np.array([[-1, -1, -1], 
                    [-1, 8, -1], 
                    [-1, -1, -1]]) 
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel2) 
plt.imshow(img)
```

```python
kernel3 = np.array([[0, -1, 0], 
                    [-1, 5, -1], 
                    [0, -1, 0]]) 
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel3) 
plt.imshow(img)
```

```python
kernel4 = np.ones((5, 5), np.float32)/30
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel4) 
plt.imshow(img)
```

```python
Gaussian = cv2.GaussianBlur(image, (15, 15), 0)
plt.imshow(Gaussian)
```

```python
from scipy import ndimage
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
k=np.ones((5, 5), np.float32)/30
img=ndimage.convolve(gray_img,k,mode="reflect")
plt.imshow(img)
```

```python
img=ndimage.convolve(gray_img,k,mode="constant")
plt.imshow(img)
```

```python
img=ndimage.convolve(gray_img,k,mode="nearest")
plt.imshow(img)
```

```python
img=ndimage.convolve(gray_img,k,mode="mirror")
plt.imshow(img)
```

```python
img=ndimage.convolve(gray_img,k,mode="wrap")
plt.imshow(img)
```
