## Image transformation using CV2

### Identity filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/Users/himanshugulechha/Desktop/MSc AI&ML/Sem3/CV/Dubai_Marina_Skyline.jpg',0)
identity_matrix = np.float32([[1, 0, 0], [0, 1, 0]])
identity_transformed = cv2.warpAffine(img, identity_matrix, (img.shape[1], img.shape[0]))
plt.imshow(identity_transformed,"gray")
```

## Image Translation

```python
rows, cols = img.shape
M = np.float32([[1, 0, 100], [0, 1, 50]])
dst = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(dst,"gray")
```

### Image Reflection

```python
M = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])
reflected_img = cv2.warpPerspective(img, M,(int(cols),int(rows)))
plt.imshow(reflected_img,"gray")
```

### Image Rotation

```python
M = np.float32([[1,  0, 0], [0, -1, rows], [0,  0, 1]])
img_rotation = cv2.warpAffine(img,cv2.getRotationMatrix2D((cols/2, rows/2),30, 0.6),(cols, rows))
plt.imshow(img_rotation,"gray")
```

### Image Scaling

```python
img_shrinked = cv2.resize(img, (250, 200),
                         interpolation=cv2.INTER_AREA)
plt.imshow(img_shrinked,"gray")
```

```python
img_enlarged = cv2.resize(img, None,
                         fx=1.5, fy=1.5,
                         interpolation=cv2.INTER_CUBIC)
plt.imshow(img_enlarged,"gray")
```

### Image Cropping

```python
cropped_img = img[100:300, 100:300]
plt.imshow(cropped_img,"gray")
```

### Image Shearing

```python
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
plt.imshow(sheared_img,"gray")
```

```python
M = np.float32([[1, 0, 0], [0.5, 1, 0], [0,   0, 1]])
sheared_img = cv2.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
plt.imshow(sheared_img,"gray")
```
