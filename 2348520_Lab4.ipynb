## Linear Spatial Filtering

### Importing the image

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
img=cv2.imread('/Users/himanshugulechha/Desktop/MSc AI&ML/Sem3/CV/Dubai_Marina_Skyline.jpg',0)
plt.imshow(img,"gray")
```

### Smoothing Linear Filter
### 1. Box filter(Mean)

```python
kernel4 = np.ones((5, 5), np.float32)/25
img_new = cv2.filter2D(src=image, ddepth=-1, kernel=kernel4) 
plt.imshow(img_new,"gray") 
plt.show()
img_3 = cv2.boxFilter(img, -25, (5,5))
plt.imshow(img_3,"gray") 
```

### 2. Gaussian Filter

```python
gaussian_blur = cv2.GaussianBlur(img,(5,5),25) 
plt.imshow(gaussian_blur,"gray") 
```

### Sharpening Filter- Mexican Hat Filter

```python
sharp_kernel = np.array([[0, -1, 0], 
                         [-1, 5, -1], 
                         [0, -1, 0]]) 
sharp_img = cv2.filter2D(src=img, ddepth=-1, kernel=sharp_kernel) 
plt.imshow(sharp_img,"gray")
```
